### ç ”ç©¶è®¡ç®—KSå€¼

```python
dat=data.loc[:,['score','od30_evv','apply_type_','month_lent']]
```

```python
def ks_go(dat):
    dat['true_label']=dat['od30_evv']
    dat['pre_prob']=dat['score']
    TPR=[]
    FPR=[]
    for i in np.arange(1000,0,-10):        
        TP = sum((dat['pre_prob'] <= i) * (dat['true_label'] == 1))   #åˆ¤ä¸ºæ­£ç±»å®žé™…ä¹Ÿä¸ºæ­£ç±»
        FP = sum((dat['pre_prob'] <= i) * (dat['true_label'] == 0))   #åˆ¤ä¸ºæ­£ç±»å®žé™…ä¸ºè´Ÿç±»  
        TN = sum((dat['pre_prob'] > i) * (dat['true_label']== 0))     #åˆ¤ä¸ºè´Ÿç±»å®žé™…ä¸ºè´Ÿç±»  
        FN = sum((dat['pre_prob'] > i) * (dat['true_label'] == 1))    #åˆ¤ä¸ºè´Ÿç±»å®žé™…ä¸ºæ­£ç±»
        TPR.append(TP/(TP+FN))
        FPR.append(FP/(FP+TN))
    ks=[]
    for i in np.arange(len(TPR)):
        ks.append(TPR[i]-FPR[i])
    return(max(ks))
```

## ä¸åŒæ”¾æ¬¾æœˆ

```python
monList=dat.drop_duplicates(subset=['month_lent'], keep='first', inplace=False).month_lent
for i in monList:
    dat1=dat[dat['month_lent']==i]
    print (i,'   KS_value:',ks_go(dat1))
```

## æ–°è€å®¢

```python
monList=dat.drop_duplicates(subset=['apply_type_'], keep='first', inplace=False).apply_type_
for i in monList:
    dat1=dat[dat['apply_type_']==i]
    print (i,'    KS_value:',ks_go(dat1))
```

```python
data=pd.read_csv('E:\\201810\\jhjj_KS\\jhjj2.csv')
dat=data.loc[:,['score','od30_od','apply_type_','month_lent']]
dat['od30_evv']=dat['od30_od']
# æŒ‰åˆ†æ•°ä»Žä½Žåˆ°é«˜æŽ’åºï¼Œå¹¶é‡è®¾ç´¢å¼•
dat=dat.sort_values('score')
dat=dat.set_index(np.arange(46472))
dat
```

```python
def ks_go3(dat):
    dat=dat.set_index(np.arange(len(dat['score'])))
    dat['true_label']=dat['od30_evv']
    dat['pre_prob']=dat['score']
    TPR=[]
    FPR=[]
    score_cut=[]
    dat_cut=dat
    sid=np.arange(0,len(dat_cut['score']),len(dat_cut['score'])/10)
    for i in sid:
        #print('å®¢æˆ·åºå·ï¼š',int(i),'   å¯¹åº”åˆ†æ•°ï¼š',dat_cut.loc[int(i),'score'])
        score_cut.append(dat_cut.loc[int(i),'score'])
    for i in score_cut:        
        TP = sum((dat['pre_prob'] <= i) * (dat['true_label'] == 1))   #åˆ¤ä¸ºæ­£ç±»å®žé™…ä¹Ÿä¸ºæ­£ç±»
        FP = sum((dat['pre_prob'] <= i) * (dat['true_label'] == 0))   #åˆ¤ä¸ºæ­£ç±»å®žé™…ä¸ºè´Ÿç±»  
        TN = sum((dat['pre_prob'] > i) * (dat['true_label']== 0))     #åˆ¤ä¸ºè´Ÿç±»å®žé™…ä¸ºè´Ÿç±»  
        FN = sum((dat['pre_prob'] > i) * (dat['true_label'] == 1))    #åˆ¤ä¸ºè´Ÿç±»å®žé™…ä¸ºæ­£ç±»
        TPR.append(TP/(TP+FN))
        FPR.append(FP/(FP+TN))
    ks=[]
    for i in np.arange(len(TPR)):
        ks.append(TPR[i]-FPR[i])
    return(max(ks))
```
